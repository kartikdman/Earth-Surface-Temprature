## **Visualizing global temperature data **

## **Dataset:** [Berkeley Earth Surface Temperature Data](https://www.kaggle.com/datasets/berkeleyearth/climate-change-earth-surface-temperature-data)

**Objective:** Analyze and visualize the global temperature data to better understand climate change and global warming trends.

**Questions:**

- How has the global average temperature changed over time?
- How do temperature anomalies vary by latitude and time?
- How do temperature trends in the Northern and Southern Hemispheres compare?
- Which countries have experienced the highest average temperature increases?
- How has the temperature changed across the globe over time?
- How effective are the visualizations in answering these questions, and what improvements can be made?

```python
#Importing necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Data\GlobalLandTemperaturesByCity.csv')
df2 = pd.read_csv('Data\GlobalLandTemperaturesByCountry.csv')
df3 = pd.read_csv('Data\GlobalLandTemperaturesByMajorCity.csv')
df4 = pd.read_csv('Data\GlobalLandTemperaturesByState.csv')
df5 = pd.read_csv('Data\GlobalTemperatures.csv')
```

### **Data Cleaning And Preprocessing Ы**

**For df (By City)**

```python
# Data preprocessing
df = df.dropna() # drop rows with missing values
df['dt'] = pd.to_datetime(df['dt']) # convert date column to datetime format
df['year'] = df['dt'].dt.year # extract year from date column
df_grouped1 = df.groupby('year')['AverageTemperature'].mean().reset_index() # group by year and calculate the average temperature 
```

**For df2 (By Country)**

```python
# Data preprocessing
df2 = df2.dropna() # drop rows with missing values
df2['dt'] = pd.to_datetime(df2['dt']) # convert date column to datetime format
df2['year'] = df2['dt'].dt.year # extract year from date column
df_grouped2 = df2.groupby('year')['AverageTemperature'].mean().reset_index() # group by year and calculate the average temperature 
```

**For df3 (By Major City)**

```python
df3 = df3.dropna() # drop rows with missing values
df3['dt'] = pd.to_datetime(df3['dt']) # convert date column to datetime format
df3['year'] = df3['dt'].dt.year # extract year from date column
df_grouped3 = df3.groupby('year')['AverageTemperature'].mean().reset_index() # group by year and calculate the average temperature
```

**For df4 (By State)**

```python
df4 = df4.dropna() # drop rows with missing values
df4['dt'] = pd.to_datetime(df4['dt']) # convert date column to datetime format
df4['year'] = df4['dt'].dt.year # extract year from date column
df_grouped4 = df4.groupby('year')['AverageTemperature'].mean().reset_index() # group by year and calculate the average temperature 
```

**For df5 (main)**

```python
df5 = df5.dropna()
df5['dt'] = pd.to_datetime(df5['dt']) # convert date column to datetime format
df5['year'] = df5['dt'].dt.year # extract year from date column
df_grouped5 = df5.groupby('year')['LandAndOceanAverageTemperature'].mean().reset_index() # group by year and calculate the average temperature
```

### **Data Visualization **

#### **How has the global average temperature changed over time?**

```python
# # Group the data by year and calculate the mean temperature for each year
global_temps_by_year = df5.groupby(df5['dt'].map(lambda x: x.year)).mean()

# Plot the global average temperature over time
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(global_temps_by_year.index, global_temps_by_year['LandAndOceanAverageTemperature'], '-o')
ax.set_xlabel('Year')
ax.set_ylabel('Global Average Temperature (掳C)')
ax.set_title('Global Average Temperature Over Time')
plt.show()
```

### **How do temperature anomalies vary by latitude and time?**

```python
# Visualize temperature anomalies by latitude and time
plt.figure(figsize=(12,5))
sns.heatmap(df.pivot_table(index='Latitude', columns='year', values='AverageTemperatureUncertainty'), cmap='coolwarm')
plt.title('Temperature Anomalies by Latitude and Time')
plt.xlabel('Year')
plt.ylabel('Latitude')
plt.show()
```

```python
df['Latitude'] = df['Latitude'].str.replace('N', '').str.replace('S', '').astype(float)
plt.figure(figsize=(12,6))
plt.hexbin(x='Latitude', y='AverageTemperature', gridsize=50, cmap='viridis', data=df)
plt.title('Temperature Anomalies by Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (Celsius)')
plt.colorbar()
plt.show()
```

#### **How do temperature anomalies vary by latitude and time?**

```python
# How has the global average temperature changed over time?
plt.figure(figsize=(12,6))
sns.lineplot(x='dt', y='LandAndOceanAverageTemperature', data=df5)
plt.title('Global Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature (Celsius)')
plt.show()
```

```python

# How do temperature trends in the Northern and Southern Hemispheres compare?
northern_df = df5[df5['LandAverageTemperature']>0]
southern_df = df5[df5['LandAverageTemperature']<0]
plt.figure(figsize=(12,6))
sns.lineplot(x='dt', y='LandAverageTemperature', data=northern_df, label='Northern Hemisphere')
sns.lineplot(x='dt', y='LandAverageTemperature', data=southern_df, label='Southern Hemisphere')
plt.title('Temperature Trends in the Northern and Southern Hemispheres')
plt.xlabel('Year')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.show()
```

```python
# Which countries have experienced the highest average temperature increases?
country_temp = df2.groupby('Country')['AverageTemperature'].mean().reset_index().sort_values(by='AverageTemperature', ascending=False).head(10)
plt.figure(figsize=(16,8))
sns.barplot(x='Country', y='AverageTemperature', data=country_temp)
plt.title('Countries with Highest Average Temperature Increases')
plt.xlabel('Country')
plt.ylabel('Temperature (Celsius)')
plt.show()
```

```python
# Group the data by latitude and year and calculate the mean temperature anomaly for each latitude and year
temps_by_lat_long_year = df.groupby([df['Latitude'].round(), df['dt'].map(lambda x: x.year)])['AverageTemperatureUncertainty'].mean().reset_index()

# Pivot the data to create a heatmap
heatmap_data = temps_by_lat_long_year.pivot(index='Latitude', columns='dt', values='AverageTemperatureUncertainty')

# Plot the heatmap
fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', center=0)
ax.set_xlabel('Year')
ax.set_ylabel('Latitude')
ax.set_title('Temperature Anomaly by Latitude and Time')
plt.show()
```

```python
# Compare temperature trends in Northern and Southern Hemispheres
north = df[df['Latitude'] >= 0].groupby('year')['AverageTemperature'].mean()
south = df[df['Latitude'] < 0].groupby('year')['AverageTemperature'].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(x=north.index, y=north.values, label='Northern Hemisphere')
sns.lineplot(x=south.index, y=south.values, label='Southern Hemisphere', color='green')
plt.title('Temperature Trends in Northern and Southern Hemispheres')
plt.xlabel('Year')
plt.ylabel('Temperature (掳C)')
plt.show()
```

```python
# Countries with the highest average temperature increases
country_avg_temp = df2.groupby('Country')['AverageTemperature'].mean().reset_index()
country_avg_temp['TemperatureChange'] = country_avg_temp['AverageTemperature'] - country_avg_temp['AverageTemperature'].shift(1)
country_avg_temp.sort_values('TemperatureChange', ascending=False, inplace=True)
top_countries = country_avg_temp.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='TemperatureChange', data=top_countries)
plt.title('Top 10 Countries with the Highest Average Temperature Increases')
plt.xlabel('Country')
plt.ylabel('Temperature Change (掳C)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Visualize temperature changes across the globe over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='dt', y='LandAndOceanAverageTemperature', data=df5)
plt.title('Temperature Changes Across the Globe Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature (掳C)')
plt.show()
```

```python
# Calculate average temperature anomaly for each country
country_temps = df2.groupby('Country')['AverageTemperature'].mean().reset_index()
country_temps['TemperatureAnomaly'] = country_temps['AverageTemperature'] - df5['LandAverageTemperature'].mean()

# Sort by temperature anomaly and display top 10
country_temps = country_temps.sort_values('TemperatureAnomaly', ascending=False)
top_10_countries = country_temps.head(10)
print(top_10_countries)
```
