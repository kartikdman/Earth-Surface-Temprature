### **Data Visualization **

#### **How has the global average temperature changed over time?**

```python
# # Group the data by year and calculate the mean temperature for each year
global_temps_by_year = df5.groupby(df5['dt'].map(lambda x: x.year)).mean()

# Plot the global average temperature over time
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(global_temps_by_year.index, global_temps_by_year['LandAndOceanAverageTemperature'], '-o')
ax.set_xlabel('Year')
ax.set_ylabel('Global Average Temperature (掳C)')
ax.set_title('Global Average Temperature Over Time')
plt.show()
```

**From the above line graph, we can see a clear trend of average temperature over time. As the years increase, the temperatures also increase. We can observe some fluctuations in the trend, but there is a noticeable difference in the year 1875, where the average temperature raised by 20-30%. From the year 1975, there is a trend with low fluctuations, and we can also see a significant increase after the year 2000. From this graph, we can conclude that there is a significant change in temperature over a period of time.**

```python
# Visualize temperature changes across the globe over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='dt', y='LandAndOceanAverageTemperature', data=df5)
plt.title('Temperature Changes Across the Globe Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature (掳C)')
plt.show()
```

### **How do temperature anomalies vary by latitude and time?**

```python
# Group the data by latitude and year and calculate the mean temperature anomaly for each latitude and year
temps_by_lat_long_year = df.groupby([df['Latitude'].round(), df['dt'].map(lambda x: x.year)])['AverageTemperatureUncertainty'].mean().reset_index()

# Pivot the data to create a heatmap
heatmap_data = temps_by_lat_long_year.pivot(index='Latitude', columns='dt', values='AverageTemperatureUncertainty')

# Plot the heatmap
fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', center=0)
ax.set_xlabel('Year')
ax.set_ylabel('Latitude')
ax.set_title('Temperature Anomaly by Latitude and Time')
plt.show()
```

**From the above Heatmap, we can observe the variations in temperature anomalies by latitude and time.**<br>
**The cooler temperatures are represented by the blue color and the hotter temperatures are represented by the red color.**<br>
**There are certain anomalies at higher latitudes, especially in the Northern Hemisphere.**<br>

```python
df['Latitude'] = df['Latitude'].str.replace('N', '').str.replace('S', '').astype(float)
plt.figure(figsize=(12,6))
plt.hexbin(x='Latitude', y='AverageTemperature', gridsize=50, cmap='viridis', data=df)
plt.title('Temperature Anomalies by Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (Celsius)')
plt.colorbar()
plt.show()
```

### **How do temperature trends in the Northern and Southern Hemispheres compare?**

```python
merged_data = df5.merge(df3, on=['dt'], how='outer')
```

```python
northern = merged_data[merged_data['AverageTemperature'] > 0]
northern['dt'] = pd.to_datetime(northern['dt'])

# Extract year from 'dt' column
northern['year'] = pd.DatetimeIndex(northern['dt']).year

plt.plot(northern['dt'], northern['AverageTemperature'])
plt.xlabel('Year')
plt.ylabel('Average Temperature (掳C)')
plt.title('Northern Hemisphere Temperature Trend')
plt.show()
southern = merged_data[merged_data['AverageTemperature'] < 0]
southern['dt'] = pd.to_datetime(southern['dt'])

# Extract year from 'dt' column
southern['year'] = pd.DatetimeIndex(southern['dt']).year


plt.plot(southern['dt'], southern['AverageTemperature'])
plt.xlabel('Year')
plt.ylabel('Average Temperature (掳C)')
plt.title('Southern Hemisphere Temperature Trend')
plt.show()
```

- **The comparison of temperature trends between the northern and southern hemispheres can be observed in the above graph.**
- **The graph clearly shows that the average temperatures in the northern hemisphere range from 0 to 30, whereas in the southern hemisphere, the temperatures range from -16 to 0.**
- **There are significant fluctuations in the temperature trends of the southern hemisphere, whereas the northern hemisphere shows a relatively stable temperature trend.**
- **Based on the analysis of these graphs, we can conclude that the northern hemisphere has experienced both the warmest and coldest temperatures in comparison to the southern hemisphere.**





### **Which countries have experienced the highest average temperature increases?**

```python
# Calculate average temperature anomaly for each country
country_temps = df2.groupby('Country')['AverageTemperature'].mean().reset_index()
country_temps['TemperatureAnomaly'] = country_temps['AverageTemperature'] - df5['LandAverageTemperature'].mean()

# Sort by temperature anomaly and display top 10
country_temps = country_temps.sort_values('TemperatureAnomaly', ascending=False)
top_10_countries = country_temps.head(10)
print(top_10_countries)
```

**Here is the top 10 countries with highest temperatures main is Djibouti**

```python
# Which countries have experienced the highest average temperature increases?
country_temp = df2.groupby('Country')['AverageTemperature'].mean().reset_index().sort_values(by='AverageTemperature', ascending=False).head(10)
plt.figure(figsize=(16,8))
sns.barplot(x='Country', y='AverageTemperature', data=country_temp)
plt.title('Countries with Highest Average Temperature Increases')
plt.xlabel('Country')
plt.ylabel('Temperature (Celsius)')
plt.show()
```

**From the above plot we can see that Djibouti is having the highest average temperature**
